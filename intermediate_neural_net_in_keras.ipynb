# Intermediate Net in KERAS

Build a intermediate neural network to classify MNIST data

#### set seed reprodicibility

```python
import numpy as np
np.random.seed(42)
```

#### Load dependencies

```python
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers
from keras.optimizers import SGD
```

#### Load Data

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

#### Preprocess Data

```python
X_train = X_train.reshape(60000, 784).astype('float32')
X_test = X_test.reshape(10000, 784).astype('float32')
```

```python
X_train /= 255
X_test /= 255
```

```python
n_classes = 10
y_train = keras.utils.to_categorical(y_train, n_classes)
y_test = keras.utils.to_categorical(y_test, n_classes)
```

#### Design neural network architecture 

```python
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(784,)))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

#### Configure Model 

```python
model.compile(loss='categorical_crossentropy', optimizer=optimizers.SGD(lr=0.1), metrics=['accuracy'])
```

#### Train

```python
model.fit(X_train, y_train, batch_size=128, epochs=200, validation_data=(X_test, y_test))
```

```python
model.evaluate(X_test, y_test)
```
